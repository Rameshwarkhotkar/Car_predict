```python
data['Current_Year']=2023
data.head(10)
```

```python
data['Total_Years']=data['Current_Year']-data['Year']
data
```

```python
data.drop(['Car_Name','Year','Current_Year'],axis=1)
```

```python
data
```

```python
car=pd.get_dummies(data,drop_first=True)
car
```

```python
sns.pairplot(car)
```

```python
axis=car.corr().index
plt.figure(figsize=(20,20))
g=sns.heatmap(car[axis].corr(),annot=True,cmap='RdYlGn')
```

```python
x=car.drop('Selling_Price',axis=1)
y=car['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
Rf_model=RandomForestRegressor()
Rf_model.fit(x_train,y_train)
```

```python
Lr_model=LinearRegression()
Lr_model.fit(x_train,y_train)
```

```python
y_pred=Lr_model.predict(x_test)
r2=r2_score(y_test,y_pred)
r2
```

```python
y_pred_train=Lr_model.predict(x_train)
r2=r2_score(y_train,y_pred_train)
r2
```

```python
y_pred=Rf_model.predict(x_test)
r2=r2_score(y_test,y_pred)
r2
```

```python
y_pred_train=Rf_model.predict(x_train)
r2=r2_score(y_train,y_pred_train)
r2
```

```python
Rf_model=RandomForestRegressor()
param_grid={'n_estimators':[50,100,200,250],'max_depth':[None,5,10,15],'min_samples_split':[2,5,10,100],'min_samples_leaf':[1,2,4]}
grid=RandomizedSearchCV(Rf_model,param_grid,cv=5)
print(grid)
grid.fit(x_train,y_train)
```
